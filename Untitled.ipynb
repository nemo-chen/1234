```python
data2['时间差'] = data2['预订任务开始时间'].apply(lambda x:int(str(x)[0:2])*60-360+int(str(x)[3:5])-30)
data2['xk']=0
for i in range(len(data2)):
    if data2['时间差'][i] == 0:
        data2.loc[i,'xk']=1/2
    else:
        data2.loc[i,'xk']=1/data2.时间差[i]
xmin = data2['xk'].min()
xmax = data2['xk'].max()
mmia = xmax-xmin
for i in range(len(data2)):
    data2.loc[i,'xk'] = (data2.loc[i,'xk']-xmin)/(mmia)
for i in range(len(data2)):
    data2.loc[i,'xk2'] = (data2.loc[i,'预订任务限额']-data2['预订任务限额'].min())/(data2['预订任务限额'].max()-data2['预订任务限额'].min())
```

```python
data2.e1 =0
data2.e2 =0
K = 1/np.log(1875)
pk1 = data2.xk.sum()
pk2 = data2.xk2.sum()
for i in range(len(data2)):
    data2.loc[i,'p1'] = data2.loc[i,'xk']/pk1
    data2.loc[i,'p2'] = data2.loc[i,'xk2']/pk2
for i in range(len(data2)):
    data2.loc[i,'e1'] = data2.loc[i,'p1']*np.log(data2.loc[i,'p1'])
    data2.loc[i,'e2'] = data2.loc[i,'p2']*np.log(data2.loc[i,'p2'])
e1 = -K*(data2['e1'][data2['e1'].notna()]).sum()
e2 = -K*(data2['e2'][data2['e2'].notna()]).sum()
r1=(1-e1)/((1-e1)+(1-e2))
r2=(1-e2)/((1-e1)+(1-e2))
for i in range(len(data2)):
    data2.loc[i,"cp"] = r1*data2.loc[i,"xk"]+r2*data2.loc[i,"xk2"]
```

```python
import geatpy as ea # import geatpy
class MyProblem(ea.Problem): # 继承Problem父类
    def __init__(self, M = 2):
        name = 'MyProblem' # 初始化name（函数名称，可以随意设置）
        Dim = len(data1)*2 # 初始化Dim（决策变量维数）
        maxormins = [1] * M # 初始化maxormins（目标最小最大化标记列表，1：最小化该目标；-1：最大化该目标）
        varTypes = [1] * Dim # 初始化varTypes（决策变量的类型，0：实数；1：整数）
        lb = [0] * Dim # 决策变量下界
        ub = [100] * Dim # 决策变量上界
        lbin = [1] * Dim # 决策变量下边界（0表示不包含该变量的下边界，1表示包含）
        ubin = [1] * Dim # 决策变量上边界（0表示不包含该变量的上边界，1表示包含）
        # 调用父类构造方法完成实例化
        ea.Problem.__init__(self, name, M, maxormins, Dim, varTypes, lb, ub, lbin, ubin)
        FieldDs = [2,lb,ub,1,0,lbin,ubin,varTypes]
    def returnf1(self):
        returnf1_ = []
        for i in range(len(data1)):
            returnf1_.append(eval('x'+str(i)))
        return returnf1_

    def returnf2(self):
        returnf2_ = []
        for i in range(len(data1)):
            returnf2_.append(eval('y'+str(i)))
        return returnf2_
    
    def f(self,money,data,i):
        money_array = np.ones(len(data))*money.min()
        z=np.array(-7.4054881-4.78421253*data.直线距离+0.1000032*money_array-0.02420188*data.预订任务限额).astype(float)
        logistic = (1/(1+np.exp(-z))).max()
        if logistic>0.5 :
            return(1)
        else :
            return(0)
        
    def sum_y(self):
        for i in range(len(data1)):
            f_data = 0
            data1_number = data1.任务号码[i]
            data_sum_y = logisticdata[logisticdata['任务号码+会员编号'].map(lambda x:x[:5])==data1_number]
            f_data = f_data+MyProblem.f(self,eval('x'+str(i)),data_sum_y,i)
            return f_data
        
    def st_list_(self):
        st=[]
        for i in range(len(logisticdata)):
            k = int(logisticdata.loc[i,'任务号码+会员编号'][1:5])-1
            s_t_ = [1/(1+np.exp(7.4054881+4.78421253*logisticdata.直线距离[i]-0.1000032*eval("x"+str(k))+\
                                0.02420188*logisticdata.预订任务限额[i]))-0.5]
            st.append(s_t_)
            st.append([1-eval("y"+str(k))])
        return st
    
    def aimFunc(self, pop): # 目标函数
        Vars = pop.Phen # 得到决策变量矩阵
        for i in range(len(data1)):
            globals()['x'+str(i)] = Vars[:, [i]]
        for i in range(len(data1)):
            globals()['y'+str(i)] = Vars[:, [i+len(data1)]]

        f1 = eval(pp)
        f2 = MyProblem.sum_y(self)
        
        list_st = MyProblem.st_list_(self)
        pop.CV = np.hstack(list_st)
        pop.ObjV = np.hstack([f1, f2]) # 把求得的目标函数值赋值给种群pop的ObjV
#        # 利用罚函数法处理约束条件
#        idx1 = np.where(x1 + x2 < 2)[0]
#        idx2 = np.where(x1 + x2 > 6)[0]
#        idx3 = np.where(x1 - x2 < -2)[0]
#        idx4 = np.where(x1 - 3*x2 > 2)[0]
#        idx5 = np.where(4 - (x3 - 3)**2 - x4 < 0)[0]
#        idx6 = np.where((x5 - 3)**2 + x4 - 4 < 0)[0]
#        exIdx = np.unique(np.hstack([idx1, idx2, idx3, idx4, idx5, idx6])) # 得到非可行解的下标
#        f1[exIdx] = f1[exIdx] + np.max(f1) - np.min(f1)
#        f2[exIdx] = f2[exIdx] + np.max(f2) - np.min(f2)
#        # 利用可行性法则处理约束条件
```

```python
import geatpy as ea # import geatpy
#from MyProblem import MyProblem # 导入自定义问题接口

if __name__ == '__main__':
    """================================实例化问题对象==========================="""
    problem = MyProblem()       # 生成问题对象
    """==================================种群设置==============================="""
    Encoding = 'BG'             # 编码方式
    NIND = 50                   # 种群规模
    Field = ea.crtfld(Encoding, problem.varTypes, problem.ranges, problem.borders) # 创建区域描述器
    population = ea.Population(Encoding, Field, NIND) # 实例化种群对象（此时种群还没被初始化，仅仅是完成种群对象的实例化）
    """================================算法参数设置============================="""
    myAlgorithm = ea.moea_NSGA2_templet(problem, population) # 实例化一个算法模板对象
    myAlgorithm.mutOper.Pm = 0.2 # 修改变异算子的变异概率
    myAlgorithm.recOper.XOVR = 0.9 # 修改交叉算子的交叉概率
    myAlgorithm.MAXGEN = 200    # 最大进化代数
    """==========================调用算法模板进行种群进化========================"""
    NDSet = myAlgorithm.run()   # 执行算法模板，得到帕累托最优解集NDSet
    NDSet.save()                # 把结果保存到文件中
    # 输出
    print('用时：%s 秒'%(myAlgorithm.passTime))
    print('非支配个体数：%s 个'%(NDSet.sizes))
    print('单位时间找到帕累托前沿点个数：%s 个'%(int(NDSet.sizes // myAlgorithm.passTime)))
```
